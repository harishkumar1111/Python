## for loop in List Comprehension

```
a= [1,2,3,4,5,6]
b=[]
for i in a:
  b.append(i+2)

print(b)
```

```
# List Comprehension
a= [1,2,3,4,5,6]
["Hi" for i in a]
```

##Conditionals in List Comprehension



```
even_numbers = []
for i in range(1,21):
  if i%2 ==0:
    even_numbers.append(i)

print(even_numbers)
```

```
# List Comprehension
[i for i in range(1,21) if i%2==0]
```

##Nested IF with List Comprehension

```
output_list = []

for i in range(1,101):
  if i%2==0:
    if i%5==0:
      output_list.append(i)

print(output_list)
```

```
# List Comprehension
[i for i in range(1,101) if i%2==0 if i%5==0 if i%50==0]
```

## if-else With List Comprehension

```
output_list = []
for i in range(1,11):
  if i%2==0:    
    output_list.append("Even")
  else:
    output_list.append("Odd")


print(output_list)
```

```
# List Comprehension
["Even" if i%2==0 else "Odd" for i in range(1,11)]
```

## if-elif-else with List Comprehension

```
l=[1,2,3]
b=[]
for v in l:
  if v == 1 :
    b.append('one')
  elif v == 2:     
    b.append('two')
  else:
    b.append('other number')

print(b)
```

```
l=[1,2,3,4]
['one' if v == 1 else 'two' if v == 2 else 'other number' for v in l]
```

## if-elif-elif-else with List Comprehension

```
l=[1,2,3,4]
b=[]
for v in l:
  if v == 1 :
    b.append('one')
  elif v == 2:     
    b.append('two')
  elif v == 3:     
    b.append('three')
  else:
    b.append('other number')

print(b)
```

```
l=[1,2,3,4]
['one' if v == 1 else 'two' if v == 2 else 'three' if v==3 else 'other number' for v in l]
```
